```python
# importing libraries
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import uniform
from scipy.stats import norm
from scipy.stats import expon
```

```python
customer_df = pd.read_csv("/Users/giulianamiranda/Documents/Labs/lab-random-variable-distributions/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv", sep = ',')

cols = []

for a in range(len(customer_df.columns)):
    cols.append(customer_df.columns[a].lower().replace(' ', '_'))
    
customer_df.columns = cols

```

```python
# Get the numerical variables from our dataset
numerical_df = customer_df.select_dtypes(include=[np.number])
```

```python
# Check using a distribution plot if the variables fit the theoretical normal or exponential distribution


def dist_plot(df):
    for column in df.columns:
        plt.figure(figsize=(8, 4))
        sns.histplot(df[column], kde=True, bins=40, color='blue', stat='density')
        plt.title(f'Distribution Plot of {column}')
        plt.xlabel(column)
        plt.ylabel('Density')
        plt.show()

dist_plot(numerical_df ) 
```

```python
# Check if any of the transformations we have seen up to this point changes the result

# log transform

log_transformed = np.log1p(numerical_df)
dist_plot(log_transformed) 
```

```python
# normalizing
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
nor_transformed = numerical_df
nor_transformed[numerical_df.columns] = scaler.fit_transform(nor_transformed[numerical_df.columns])

dist_plot(nor_transformed) 
```

```python
# Standardized 

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

stand_transformed = numerical_df
stand_transformed[numerical_df.columns] = scaler.fit_transform(stand_transformed[numerical_df.columns])

dist_plot(stand_transformed) 
```

```python
# Yes, applying the transformations can afect the distribution, but the effects may vary a lot. 
# A good example of a high variance is the 'months_since_last_claim'
```
