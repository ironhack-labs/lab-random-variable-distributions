```python
# Importing libraries
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
# Importing data
df_customer = pd.read_csv(r"C:\Users\joaoa\Desktop\Ironhack\Downloads_Ironhack\we_fn_use_c_marketing_customer_value_analysis.csv")
df_customer
```

```python
# Standardizing columns names
cols = []
for i in range(len(df_customer.columns)):
    cols.append(df_customer.columns[i].lower().replace(" ","_"))
df_customer.columns = cols
df_customer
```

```python
# Checking data types
df_customer.dtypes
```

```python
# 1) Getting the numerical variables from the dataset
df_numerical = df_customer.select_dtypes(include = np.number)
df_numerical
```

```python
# 2) Checking if the variables fit the theoretical normal or exponential distribution
# From the distribution plots shown below, no one fit the theoretical normal distribution (bell-shaped function)
# However, the 'month_since_plicy_inception' distribution is the closest to a normal distribution.
# Actually, the 'customer_lifetime_value' distribution could be exponential, but we needed to check with a proper test.

for col in df_numerical.columns:
    sns.distplot(df_numerical[col])
    plt.show()
```

```python
# 3.a) Checking if normalization changes the result
# The conclusions stated in 2) didn't change.

scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df_numerical)
normalized_data = pd.DataFrame(normalized_data, columns = df_numerical.columns)

for col in normalized_data.columns:
    sns.distplot(normalized_data[col])
    plt.show()
```

```python
# 3.b) Checking if standardization changes the result
# The conclusions stated in 2) didn't change.

scaler = StandardScaler()
stand_data = scaler.fit_transform(df_numerical)
stand_data = pd.DataFrame(stand_data, columns = df_numerical.columns)

for col in stand_data.columns:
    sns.distplot(stand_data[col])
    plt.show()
```

```python
# 3.c) Checking if log transformation changes the result
# We didn't learn how to do a log transformation. 
# Despite that I tried, but I don't know how to plot a "-infinite" value. 
# I could give a super negative number but, depending on that number, I would have different outputs.

log_data = df_numerical
for col in log_data.columns:
    log_data[col] = np.log10(log_data[col])
log_data
```
