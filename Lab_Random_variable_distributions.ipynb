```python

import pandas as pd
import numpy as np
```

```python
customer_df = pd.read_csv(r'C:\Users\Tracy\Desktop\Class1IR\Cleaning_categorical_Data_Python\we_fn_use_c_marketing_customer_value_analysis.csv')
customer_df
```

```python
# renamings the columns so they follow the standart 

cols = []

for i in range(len(customer_df.columns)) :
    cols.append(customer_df.columns[i].lower().replace(' ','_'))
customer_df.columns = cols

customer_df
```

```python
# find all the Numerical Data and create a New Data Frame

Numerical_df = customer_df.select_dtypes(include = 'number')

Numerical_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_num(Numerical_df):
    for column in Numerical_df.columns:
        sns.histplot(x=Numerical_df[column])
        plt.show()
    return None
   

plot_num(Numerical_df)


# 2) Checking if the variables fit the theoretical normal or exponential distribution
# From the distribution plots shown below, no one fit the theoretical normal distribution (bell-shaped function)
# 'month_since_plicy_inception' distribution is the closest to a normal distribution.
#The 'customer_lifetime_value' distribution could be exponential, but I am not sure about it 
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# 3.a) Checking if normalization changes the result
# The conclusions stated in 2) didn't change.

scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(Numerical_df)
normalized_data = pd.DataFrame(normalized_data, columns = Numerical_df.columns)

for col in normalized_data.columns:
    sns.distplot(normalized_data[col])
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# 3.b) Checking if standardization changes the result
# The conclusions stated in 2) didn't change.

scaler = StandardScaler()
stand_data = scaler.fit_transform(Numerical_df)
stand_data = pd.DataFrame(stand_data, columns = Numerical_df.columns)

for col in stand_data.columns:
    sns.distplot(stand_data[col])
    plt.show()
```
