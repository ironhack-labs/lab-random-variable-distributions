```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',None)
```

```python
data = pd.read_csv('marketing.csv')
data.shape
```

```python
data.head(10)
```

```python
numerical = data.select_dtypes(np.number)
numerical.head()
```

```python
print(numerical.columns)
```

```python
for column in numerical.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=numerical, x=column, kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Apply log-transform to the numerical columns
numerical_log_transformed = numerical.apply(lambda x: x.apply(lambda x: x if x <= 0 else np.log(x)))

# Plot distribution plots for each log-transformed numerical column
for column in numerical_log_transformed.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=numerical_log_transformed, x=column, kde=True)
    plt.title(f'Distribution of Log-Transformed {column}')
    plt.xlabel(f'Log({column})')
    plt.ylabel('Frequency')
    plt.show()
```
