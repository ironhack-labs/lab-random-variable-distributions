```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv', index_col=0)
data.head()
```

```python
#Get the numerical variables from our dataset.
```

```python
numeric = data.select_dtypes(include=[int, float])
numeric.head()
```

```python
#Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
for col in numeric:
    sns.distplot(numeric[col], kde=True, hist=False)
    plt.title(f'Distribution of {col}')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
```

```python
# Some are normal, some are exponential.
```

```python
#Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
numeric_log = numeric.apply(lambda x: np.log(x))
numeric_log.head()
```

```python
for col in numeric_log:
    sns.distplot(numeric_log[col], kde=True, hist=False)
    plt.title(f'Distribution of {col}')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
```

```python
# Yes, they changed.
```
