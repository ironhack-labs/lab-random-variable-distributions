```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import uniform
from scipy.stats import norm
from scipy.stats import expon

# Matplot
import matplotlib.pyplot as plt
%matplotlib inline

# Seaborn
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv (r"C:\Users\filip\OneDrive\Desktop\IRONHACK\Labs\Week10\lab-random-variable-distributions\files_for_lab\we_fn_use_c_marketing_customer_value_analysis.csv")

cols = []
for i in range(len(df.columns)): 
    cols.append(df.columns[i].lower().replace(' ', '_')) 
df.columns = cols
```

```python
# 1. Get the numerical variables from our dataset.

df_num = df.select_dtypes(exclude=['object'])

df_num.isna().sum().sum()

df_num.dtypes
```

```python
# 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

# 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.


scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df_num)
normalized_data = pd.DataFrame(normalized_data, columns=df_num.columns)

log_transformed = np.log1p(df_num)
log_data = scaler.fit_transform(log_transformed)
log_data = pd.DataFrame(log_data, columns = log_transformed.columns)

plt.xlabel('customer_lifetime_value  dist')
plt.ylabel('Frequency')
plt.title('customer_lifetime_value  dist')
sns.distplot(log_data['customer_lifetime_value'])
plt.show()

plt.xlabel('customer_lifetime_value  dist')
plt.ylabel('Frequency')
plt.title('customer_lifetime_value  dist')
sns.distplot(normalized_data['customer_lifetime_value'])
plt.show()
```
