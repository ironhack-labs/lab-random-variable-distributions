```python
import imblearn
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import os
from math import ceil 
import warnings
warnings.filterwarnings("ignore")
```

```python
marketing = pd.read_csv("/Users/giacomorossini/Desktop/IRONHACK/Laboratory/LAB 15/LAB_random_var_dist/lab-random-variable-distributions/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
marketing_num = marketing.select_dtypes(include = np.number)
```

```python
def stacked_distplot(df):
    sns.set()

    fig, axes=plt.subplots(2, ceil(len(df.columns)/2),figsize=(20,11))

    for ax, feat in zip(axes.flatten(), df.columns): 
        sns.distplot(df[feat], ax=ax)
        ax.set_title(feat, y=-0.14)
        
    title = "Distribution"
    plt.suptitle(title)
    
    plt.show()
```

```python
stacked_distplot(marketing_num)
```

```python
from sklearn.preprocessing import StandardScaler
scaler1 = StandardScaler() #This is called initialization
stand_data = scaler1.fit_transform(marketing_num) # Fit into data
stand_data = pd.DataFrame(stand_data, columns= marketing_num.columns)
```

```python
stacked_distplot(stand_data)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() 
normalized_data = scaler.fit_transform(marketing_num) 
normalized_data = pd.DataFrame(normalized_data, columns= marketing_num.columns)
```

```python
stacked_distplot(normalized_data)
```

```python
def subplot_distplots(dataframe):
    sns.set()
    fig, axes=plt.subplots(2, ceil(len(dataframe.columns)/2),figsize=(20,11))
    

    # Plot distplots for each numerical column
    
    for ax, feat in zip(axes.flatten(), dataframe.columns):
        sns.histplot(dataframe[feat], kde=True, ax=ax)
        ax.set_title(feat, y=-0.14)
    
    title = "Continuos Variables"
    plt.suptitle(title)    

    # Show the plots
    plt.show()

# Example usage:
# Assuming 'df' is your DataFrame
subplot_distplots(num_log)
```
